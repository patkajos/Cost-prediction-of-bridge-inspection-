```python
def clean(X):
    for i in list(X.columns):
        if X[i].isnull().sum().sum() > X.shape[0]-X.shape[0]*0.5:
            X.drop(columns = i , inplace = True)
    return X
```

```python
clean(df)
```

```python
df.dropna(inplace = True)
```
